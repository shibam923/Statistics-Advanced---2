Question 1: What is hypothesis testing in statistics? 

Answer:  Hypothesis testing in statistics is a **method used to make decisions or inferences about a population based on sample data**.

It involves checking whether there is enough statistical evidence in the sample to support or reject a claim (called a *hypothesis*) about the population.

### Key Concepts:

1. **Null Hypothesis (H₀):**
   The default assumption or claim that there is *no effect*, *no difference*, or *status quo* (e.g., “the average test score is 70”).

2. **Alternative Hypothesis (H₁ or Ha):**
   The competing claim that we want to test (e.g., “the average test score is not 70”).

3. **Test Statistic:**
   A standardized value (like a *z-score* or *t-score*) calculated from the sample data to assess how far it deviates from what is expected under H₀.

4. **P-value:**
   The probability of observing the data (or something more extreme) if the null hypothesis is true.

   * Small p-value (≤ significance level, α) → reject H₀.
   * Large p-value (> α) → fail to reject H₀.

5. **Significance Level (α):**
   The threshold (commonly 0.05) used to decide whether to reject H₀.

---

✅ **In short:** Hypothesis testing is a decision-making process that helps determine if sample evidence is strong enough to reject a null hypothesis in favor of an alternative.

Question 3: Explain the significance level in hypothesis testing and its role in deciding 
the outcome of a test. 

Answer: 

---

### **Significance Level (α):**

* The **significance level**, denoted by **α (alpha)**, is the **threshold probability** we set **before** conducting a hypothesis test.
* It represents the **maximum risk of rejecting the null hypothesis (H₀) when it is actually true** (a *Type I error*).

Common choices: **α = 0.05 (5%)**, **α = 0.01 (1%)**, **α = 0.10 (10%)**.

---

### **Role in Hypothesis Testing:**

1. **Compare p-value with α:**

   * If **p-value ≤ α** → Reject H₀ (evidence supports H₁).
   * If **p-value > α** → Fail to reject H₀ (not enough evidence against H₀).

2. **Controls Error Risk:**

   * By setting α, we decide how strict we want to be in avoiding a **false positive** (rejecting a true H₀).
   * Example: With α = 0.05, we accept a **5% chance of being wrong** when rejecting H₀.

---

### **Example:**

Suppose a company claims the average life of its batteries is **100 hours**.

* **H₀:** μ = 100
* **H₁:** μ ≠ 100
* Significance level: **α = 0.05**

If the test produces a **p-value = 0.03**, since **0.03 < 0.05**, we **reject H₀** and conclude that the average battery life is significantly different from 100 hours.

---

✅ **In short:**
The **significance level (α)** is like a **cutoff line** that helps us decide whether sample evidence is strong enough to reject the null hypothesis.

Question 4: What are Type I and Type II errors? Give examples of each. 

Answer 

---

### **1. Type I Error (False Positive):**

* **Definition:** Rejecting the null hypothesis (H₀) when it is actually true.
* It’s like concluding there *is an effect* when in reality there is **none**.
* Probability of making this error = **α (significance level)**.

✅ **Example:**
A medical test for a disease says a healthy person **has the disease** (rejects H₀: “patient is healthy” when H₀ is actually true).

---

### **2. Type II Error (False Negative):**

* **Definition:** Failing to reject the null hypothesis (H₀) when it is actually false.
* It’s like concluding there *is no effect* when in reality there **is one**.
* Probability of making this error = **β** (related to test *power = 1 – β*).

✅ **Example:**
A medical test fails to detect a disease in a sick person (fails to reject H₀: “patient is healthy” even though H₀ is false).

---

### **Analogy (Courtroom Example):**

* **Type I Error:** Convicting an innocent person.
* **Type II Error:** Letting a guilty person go free.

---

Question 5: What is the difference between a Z-test and a T-test? Explain when to use 
each.
 
Answer

---

## **Z-test vs T-test**

### **1. Z-test**

* **Used when:**

  * The population variance (σ²) is **known**.
  * The sample size is **large (n > 30)** (Central Limit Theorem makes the sampling distribution approximately normal).
* **Distribution Used:** Standard **Normal distribution (Z-distribution)**.
* **Example:** Testing if the average height of students is 170 cm when the population standard deviation is known.

---

### **2. T-test**

* **Used when:**

  * The population variance (σ²) is **unknown**.
  * The sample size is **small (n ≤ 30)**.
  * Sample is drawn from a population that is approximately normal.
* **Distribution Used:** **Student’s t-distribution**, which is wider and accounts for extra uncertainty when σ is unknown.
* **Example:** Testing if the average score of 15 students differs from 50, when population variance is unknown.

---

### **Key Differences (Quick Table):**

| Feature             | **Z-test**                   | **T-test**                 |
| ------------------- | ---------------------------- | -------------------------- |
| Population variance | Known                        | Unknown                    |
| Sample size         | Large (n > 30)               | Small (n ≤ 30)             |
| Distribution used   | Normal (Z)                   | Student’s t                |
| Shape               | Narrow (less uncertainty)    | Wider (more uncertainty)   |
| Example             | Quality control with known σ | Small sample medical study |

---

✅ **In short:**

* Use a **Z-test** when σ is known or n is large.

Question 8: Write a Python script to simulate data from a normal distribution and 
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. 
(Include your Python code and output in the code box below.) 
    
Answer:  We’ll simulate data from a normal distribution, compute the 95% confidence interval (CI) for the mean, and then plot the data using Matplotlib.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
std_error = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval using t-distribution
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

margin_of_error = t_critical * std_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample mean:", sample_mean)
print("95% Confidence Interval: [{:.2f}, {:.2f}]".format(ci_lower, ci_upper))

# Step 4: Plot the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
````
